```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
value=os.getenv("groq_api_key")

if not value:
    raise EnvironmentError("GROQ_API_KEY not set")
os.environ["groq_api_key"]=value
```

```python
from langchain.chat_models import init_chat_model

model=init_chat_model("groq:qwen/qwen3-32b")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import ChatPromptTemplate

def story_analyser():
    story_prompt=ChatPromptTemplate([
        ("system","You are a creative story creator and ypou hava to create a story using the user input."),
        ("user","Theme:{theme},Main character:{chararcter} and setting:{setting}.")
    ])

    analysis_prompt=ChatPromptTemplate([
        ("system","You are a litracy critic.Your job is to analyse this story and respond professionally."),
        ("user","{story}")
    ])

    create_story_chain=(
        story_prompt
        | model
        | StrOutputParser()
    )


    def analyzerPromptSetter(story_text):
        return {"story":story_text}
    
    analyze_chain=(
        create_story_chain
        | RunnableLambda(analyzerPromptSetter)
        | analysis_prompt
        | model
        | StrOutputParser()
    )

    return analyze_chain
```

```python
chain=story_analyser()
```

```python
inputs={
    "theme":"Cloud Computing",
    "chararcter":"A cloud based AI agent",
    "setting":"memeory space or hard disk of a server."
}

for chunk in chain.stream(inputs):
    print(chunk,end="",flush=True)
```
